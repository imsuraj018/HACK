```python
import json
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image


MODEL_PATH = "maize_disease_model.h5"  # your trained model
RECOMM_JSON = "Maize_disease_recommendations1.json"  # JSON file we created
IMG_PATH = "/home/sk/Desktop/crop_disease_dataset/Maize_dataset/data/Blight/Corn_Blight (1).jpeg"  # image to predict


model = load_model(MODEL_PATH)


with open(RECOMM_JSON, "r") as f:
    recommendations = json.load(f)
class_names = [
    "Maize_Northern_Leaf_Blight",
    "Maize Common Rust",
    "Maize_Gray_Leaf_Spot",
    "Healthy",
    
  
]


img = image.load_img(IMG_PATH, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = tf.keras.applications.efficientnet_v2.preprocess_input(img_array)

pred_probs = model.predict(img_array)
pred_index = np.argmax(pred_probs, axis=1)[0]
pred_class = class_names[pred_index]


advice = recommendations.get(pred_class, {"Spray": "No info", "Fertilizer": "No info"})

print(f"Predicted Disease: {pred_class}")
advice_all_languages = recommendations[pred_class]

for language, advice in advice_all_languages.items():
    print(f"Language: {language}")
    print(f"  Recommended Spray: {advice['description']}")
    print(f"  Recommended Fertilizer: {advice['recommendation']}\n")
```
